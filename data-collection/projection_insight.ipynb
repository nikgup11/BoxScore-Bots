```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('./output_data/linear_player_prediction.csv')
#df = pd.read_csv('./output_data/tonights_projections_rnn.csv')
```

```python
average_mse = df['MSE'].mean()
median_mse = df['MSE'].median()
print('median MSE: ' + str(median_mse))
print('average MSE: ' + str(average_mse))
below_average_mse = df[(df['MSE'] < average_mse) & (df['predictedPoints'] > 10)]
below_median_mse = df[(df['MSE'] < median_mse) & (df['predictedPoints'] > 10)]
print('Players with below average MSE: ' + str(len(below_average_mse)))
print('Players with below median MSE: ' + str(len(below_median_mse)))
below_median_mse.sort_values(by=['MSE'], ascending=True, inplace=True)
below_median_mse.head(20)
```

```python
df.sort_values(by=['predictedPoints'], ascending=False, inplace=True)
df.head(10)
```

```python
df = df[df['predictedPoints'] > 10]
df.sort_values(by=['MSE'], ascending=True, inplace=True)
df.head(20)
```

```python
plt.scatter(df['predictedPoints'], df['MSE'])
plt.xlabel('Predicted Points')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Predicted Points vs MSE')
for idx, row in df.iterrows():
    plt.annotate(f"{row['firstName']} {row['lastName']}", 
                 (row['predictedPoints'], row['MSE']),
                 fontsize=8, alpha=0.7)
plt.show()
```

```python
tracking = pd.read_csv('./output_data/projection_tracking.csv')
tracking.sort_values(by=['Game_Date'], ascending=True, inplace=True)
```

```python
tracking_clean = tracking.dropna(subset=['Actual_PTS'])
close_projections = tracking_clean[(abs(tracking_clean['Proj_PTS'] - tracking_clean['Actual_PTS']) <= 3) & (tracking_clean['Actual_PTS'] != 0) & (tracking_clean['Proj_PTS'] != 0)]
close_projections[['Name', 'Proj_PTS', 'Actual_PTS']]
```

```python
# on average how far off are the projections?
# if < 1 model is doing well
average_dist = tracking_clean['Actual_PTS'] - tracking_clean['Proj_PTS']
average_dist.mean()
```

```python
rnn_projections = pd.read_csv('./output_data/tonights_projections_rnn.csv')
rnn_projections.sort_values(by=['Total_PRA'], ascending=False, inplace=True)
rnn_projections.head(20)
```

```python
unique_dates = tracking['Game_Date'].unique()
dist_by_date = {}
for date in unique_dates:
    dist_by_date[date] = average_dist[tracking_clean['Game_Date'] == date].mean()

for day in dist_by_date:
    print(f"{day}: {dist_by_date[day]}")
```
